## E-Learning Company
Goal: Find out the two best markets to adverstise our product in

We will be using the freeCodeCamp's 2017 New Coder Suvery dataset for this study.
The data is publicly available in [this github repository](https://github.com/freeCodeCamp/2017-new-coder-survey)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("2017-fCC-New-Coders-Survey-Data.csv",low_memory=False)
data.head()
```

```python
rows, cols = data.shape
print("There are {} rows and {} columns".format(rows, cols))
```

```python
list(data.columns)
```

```python
job_interest_freq = data['JobRoleInterest'].value_counts(normalize = True) * 100
```

```python
job_interest_freq.sort_values(ascending = False)
```

People are mostly interested in Full-Stack Web Developer,Front-End Web Developer and Data Scientist.
We can also see that people are interested inmore than one fields

```python
interest = data['JobRoleInterest'].dropna(axis = 0)
interest_list = interest.str.split(',')
```

```python
interest_count = interest_list.apply(lambda x: len(x))
interest_count.value_counts(normalize = True) * 100
```

Almost 31% of people are interested in single subject only while less than 1 percent of people are interest in more than 9 subjects

Since we are focusing on web or mobile development

```python
web_mobile = interest.str.contains('Web Developer|Mobile Developer')
```

```python
freq = web_mobile.value_counts(normalize = True) * 100
```

```python
freq.plot.bar()
plt.title("Participants interest in web or mobile development")
plt.ylabel("Percentage")
plt.xticks([0,1], ['Web or Mobile development', 'Other subject'], rotation = 45)
plt.show()
```

Around 86% of people are interested in web or mobile development

```python
data_clean = data[data['JobRoleInterest'].notnull()].copy()

absolute_freq = data_clean['CountryLive'].value_counts()
relative_freq = data_clean['CountryLive'].value_counts(normalize = True) * 100

pd.DataFrame({"absolute freq": absolute_freq, "percentage" : relative_freq})
```

45% of the people are currenlty from USA. While India has the second largest customer. 

It also seems like a good idea to narrow down our analysis to only four countries: the US, India, the United Kingdom, and Canada. Two reasons for this decision are:

These are the countries having the highest absolute frequencies in our sample, which means we have a decent amount of data for each.
Our courses are written in English, and English is an official language in all these four countries. The more people that know English, the better our chances to target the right people with our ads.

```python
data_clean['MonthsProgramming'].unique()
```

Some students answered that they had been learning to code for 0 months (it might be that they had just started when they completed the survey). 

```python
data_clean['MonthsProgramming'].replace(0,1, inplace = True)

data_clean['money_per_month'] = data_clean['MoneyForLearning']/data_clean['MonthsProgramming']
```

```python
data_clean['money_per_month'].isna().value_counts(normalize = True)
```

So around 90% of the customers are spending money for learning

```python
data_clean = data_clean[data_clean['money_per_month'].notnull()]
data_clean = data_clean[data_clean['CountryLive'].notnull()]
```

Since we are concerned with the amount spend by customers we remove records where people didnt spendmoney

```python
money_spent_per_country = data_clean.groupby('CountryLive')['money_per_month'].agg(np.mean)
```

We are interest in top 4 countries that spent most money. 

```python
money_spent_per_country[['United States of America',
                            'India', 'United Kingdom',
                            'Canada']]
```

Though Canda has less number of customers in comparison to United Kingdom, the amount of money spend is more for Canada. Based on customer of India, they should have more money spend but the money spend by India and Canda is on par. There could be various socio-economic factor contributing to this.

```python
top_4 = data_clean[data_clean['CountryLive'].str.contains(
    'United States of America|India|United Kingdom|Canada')]
import seaborn as sns
sns.boxplot(x = 'CountryLive', y = 'money_per_month', data= top_4)
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada'])
plt.title("Money Spent per Country")
plt.ylabel("Money per Month ($)")
plt.xlabel("Country")
plt.show()
```

We are unable to see the boxes, but we can see that 2 person have spent more than $50k. This is generally hard to belive so we will remove this record

```python
data_clean = data_clean[data_clean['money_per_month'] < 20000]
data_clean.groupby('CountryLive')['money_per_month'].mean()[['United States of America',
                            'India', 'United Kingdom',
                            'Canada']]
```

```python
top_4 = data_clean[data_clean['CountryLive'].str.contains(
    'United States of America|India|United Kingdom|Canada')]

sns.boxplot(x = 'CountryLive', y = 'money_per_month', data=top_4)
plt.title("Money spend each month by country")
plt.xlabel("Country")
plt.ylabel("Money spent per country ($)")
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) # avoids tick labels overlap
plt.show()
```

We can see a few extreme outliers for India (values over $2500 per month), but it's unclear whether this is good data or not. Maybe these persons attended several bootcamps, which tend to be very expensive. Let's examine these two data points to see if we can find anything relevant.

```python
india_outliers = data_clean[(data_clean['CountryLive'] == 'India') & (data_clean['money_per_month'] >= 2500)]
india_outliers
```

None of the people from India actually attended the bootcamp so we cannot be sure where the amount is coming from. So we will treat them as outliers

```python
top_4 = top_4.drop(india_outliers.index)
```

We will check for USA for above 6000

```python
usa_outliers = data_clean[(data_clean['CountryLive'] == 'United States of America') & (data_clean['money_per_month'] >= 6000)]
usa_outliers
```

Some of the people didnt attend the bootcamp so we will remove them as well

```python
top_4 = top_4.drop(usa_outliers[usa_outliers['AttendedBootcamp'] == 0.0].index)
```

We will look for Canada for money spend above 4000

```python
canada_outliers = data_clean[(data_clean['CountryLive'] == 'Canada') & (data_clean['money_per_month'] >= 4000)]
canada_outliers
```

The person has attended the bootcamp, s we wont remove this record

```python
top_4.groupby('CountryLive').mean()['money_per_month']
```

```python
sns.boxplot(x = 'CountryLive', y = 'money_per_month', data=top_4)
plt.title("Money spend each month by country")
plt.xlabel("Country")
plt.ylabel("Money spent per country ($)")
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) # avoids tick labels overlap
plt.show()
```

We should obviously market in the USA but we should also market inthe Canada for the subscription at a price of $59 per month. India spends around $65 per month, so we can also market in the India as well. But same thing cannot be said for the UK

```python
top_4['CountryLive'].value_counts(normalize = True) * 100
```

It seems more tempting to spend in Canada then India due to various socio-economic factors.

We should spend the budget unequally that is more than 60% in USA, 25% in Canada and 15% in India. Other options can also be choosen but irrespective of the way, more amount should be allocated for USA, then Canada and India.

We could only advertise only in USA but since we want to grow across the world it would not make sense for long run.
